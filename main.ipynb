### Simulated Annealing

#### Parameters

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
# Set the style for professional appearance
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("m" \
"uted")


# my version
df_benchmark = pd.read_csv("D:\\leona\\University\\Third Year\\Thesis\\Code\\thesis\\data\\rsa\\extra_benchmark.csv")
df_exploration = pd.read_csv("D:\\leona\\University\\Third Year\\Thesis\\Code\\thesis\\data\\rsa\\extra_exploration.csv")
df_optimization = pd.read_csv("D:\\leona\\University\\Third Year\\Thesis\\Code\\thesis\\data\\rsa\\extra_optimization.csv")
df_narrow = pd.read_csv("D:\\leona\\University\\Third Year\\Thesis\\Code\\thesis\\data\\rsa\\extra_narrow.csv")
df_wide = pd.read_csv("D:\\leona\\University\\Third Year\\Thesis\\Code\\thesis\\data\\rsa\\extra_wide.csv")


df_benchmark["type"] = "benchmark"
df_exploration["type"] = "exploration"
df_optimization["type"] = "optimization"
df_all = pd.concat([
    df_benchmark,
    df_exploration,
    df_optimization
], ignore_index=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create figure with shared x-axis
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8), sharex=True)

# Plot 1 (top): Cost vs Iteration
sns.lineplot(data=df_all, x="iteration", y="cost", ax=ax1,
             linewidth=2, hue="type")
ax1.set_ylabel('Cost', fontsize=12, fontweight='semibold', labelpad=15)
ax1.grid(True, alpha=0.3)
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
ax1.set_xlim(0, 1000)
ax1.set_ylim(-1, 30.0)


# Plot 2 (bottom): Frequency vs Iteration
sns.lineplot(data=df_all, x="iteration", y="freq", ax=ax2,
             linewidth=1, hue="type")
ax2.set_xlabel('Iteration', fontsize=12, fontweight='semibold', labelpad=15)
ax2.set_ylabel('Frequency', fontsize=12, fontweight='semibold', labelpad=15)
ax2.grid(True, alpha=0.3)
ax2.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)
ax2.set_xlim(-10, 1000)
ax2.set_ylim(-0.03, 1.0)

# Remove the second legend and place one legend at the top-right from the first axes
handles, labels = ax1.get_legend_handles_labels()
# Remove the second legend
ax2.get_legend().remove()

# Custom legend settings for academic style
handles, labels = ax1.get_legend_handles_labels()
legend = ax1.legend(handles, labels, title="Type",
                    loc='upper left', bbox_to_anchor=(1.02, 1),
                    frameon=False, fontsize=11, title_fontsize=12)

# Optional: Set serif font (global)
plt.rcParams['font.family'] = 'serif'

# Tidy layout
plt.tight_layout(pad=2.0)
fig.patch.set_facecolor('white')
plt.show()
```

```python
import seaborn as sns
import pandas as pd

df = pd.read_csv("D:\\leona\\University\\Third Year\\Thesis\\Code\\thesis\\data\\rgd\\comparison_size_data.csv")
df_alpha = pd.read_csv("D:\\leona\\University\\Third Year\\Thesis\\Code\\thesis\\data\\rgd\\comparison_alpha_data.csv")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set a professional style
sns.set(style="whitegrid", context="talk", palette="deep")

# Create a high-resolution figure
plt.figure(figsize=(10, 6), dpi=300)

# Point plot with improved formatting
ax = sns.pointplot(
    data=df,
    x="size",
    y="epochs",
    hue="type",
    dodge=True,
    markers="o",
    linestyles="-",
    errwidth=1.5,
    capsize=0.1
)

# Axis labels with units (if applicable)
ax.set_xlabel("Model Size", fontsize=14)
ax.set_ylabel("Training Epochs", fontsize=14)

# Optional: add a title
ax.set_title("Training Epochs by Model Size and Type", fontsize=16, weight="bold")

# Improve legend formatting
ax.legend(title="Model Type", title_fontsize=12, fontsize=11)

# Optional: rotate x-axis labels if crowded
# plt.xticks(rotation=45)

# Tight layout for clean output
plt.tight_layout()

# Show or save
plt.show()
# plt.savefig("pointplot_epochs_by_size.png", bbox_inches="tight")
```
