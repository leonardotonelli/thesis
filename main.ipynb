```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
# Set the style for professional appearance
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("husl")


# my version
df_benchmark = pd.read_csv("D:\\leona\\University\\Third Year\\Thesis\\Code\\thesis\\data\\rsa\\extra_benchmark.csv")
df_exploration = pd.read_csv("D:\\leona\\University\\Third Year\\Thesis\\Code\\thesis\\data\\rsa\\extra_exploration.csv")
df_optimization = pd.read_csv("D:\\leona\\University\\Third Year\\Thesis\\Code\\thesis\\data\\rsa\\extra_optimization.csv")
df_narrow = pd.read_csv("D:\\leona\\University\\Third Year\\Thesis\\Code\\thesis\\data\\rsa\\extra_narrow.csv")
df_wide = pd.read_csv("D:\\leona\\University\\Third Year\\Thesis\\Code\\thesis\\data\\rsa\\extra_wide.csv")
```

```python
# Create figure with subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Plot 1: Frequency over iterations
sns.lineplot(data=df_benchmark, x="iteration", y="freq", ax=ax1, linewidth=2, marker='.', markersize=2)
ax1.set_title('Frequency vs Iteration', fontsize=14, fontweight='bold', pad=20)
ax1.set_xlabel('Iteration', fontsize=12, fontweight='semibold')
ax1.set_ylabel('Frequency', fontsize=12, fontweight='semibold')
ax1.grid(True, alpha=0.3)
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)

# Plot 2: Cost over iterations
sns.lineplot(data=df_benchmark, x="iteration", y="cost", ax=ax2, linewidth=2.5, marker='.', markersize=1)
ax2.set_title('Cost vs Iteration', fontsize=14, fontweight='bold', pad=20)
ax2.set_xlabel('Iteration', fontsize=12, fontweight='semibold')
ax2.set_ylabel('Cost', fontsize=12, fontweight='semibold')
ax2.grid(True, alpha=0.3)
ax2.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)

# Adjust layout and spacing
plt.tight_layout(pad=3.0)

# Optional: Add subtle background color
fig.patch.set_facecolor('white')

# Display the plot
plt.show()
```

### Gradient Descent


```python
import seaborn as sns
import pandas as pd

df = pd.read_csv("D:\\leona\\University\\Third Year\\Thesis\\Code\\thesis\\data\\rgd\\comparison_size_data.csv")
df_alpha = pd.read_csv("D:\\leona\\University\\Third Year\\Thesis\\Code\\thesis\\data\\rgd\\comparison_alpha_data.csv")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set a professional style
sns.set(style="whitegrid", context="talk", palette="deep")

# Create a high-resolution figure
plt.figure(figsize=(10, 6), dpi=300)

# Point plot with improved formatting
ax = sns.pointplot(
    data=df,
    x="size",
    y="epochs",
    hue="type",
    dodge=True,
    markers="o",
    linestyles="-",
    errwidth=1.5,
    capsize=0.1
)

# Axis labels with units (if applicable)
ax.set_xlabel("Model Size", fontsize=14)
ax.set_ylabel("Training Epochs", fontsize=14)

# Optional: add a title
ax.set_title("Training Epochs by Model Size and Type", fontsize=16, weight="bold")

# Improve legend formatting
ax.legend(title="Model Type", title_fontsize=12, fontsize=11)

# Optional: rotate x-axis labels if crowded
# plt.xticks(rotation=45)

# Tight layout for clean output
plt.tight_layout()

# Show or save
plt.show()
# plt.savefig("pointplot_epochs_by_size.png", bbox_inches="tight")
```
